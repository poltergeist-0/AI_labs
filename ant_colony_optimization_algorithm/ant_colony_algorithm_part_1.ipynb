```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def euclidean_distance(x1,y1, x2,y2):
    return ((x1-x2)**2 + (y1-y2)**2 )**0.5
```

```python
n = random.randint(25,35)
print(n)
# n = 10

x = np.linspace(0,200,200)

l_x = [100]
l_y = [100]
l = [[100,100]]

for i in range(n-1):
    x,y = -1,-1
    flag = 0
    while flag == 0:
        x,y = random.randint(0,200),random.randint(0,200)
        flag = 1
        for j in range(len(l_x)):
            dist = euclidean_distance(l_x[j],l_y[j], x,y)
            if dist < 10 or dist > 100: flag = 0
    l_x.append(x)
    l_y.append(y)
    l.append([x,y])

plt.plot(l_x,l_y,"ok",markersize=4)
plt.show()
```

```python
f = open("for_ant_colony_algorithm.txt", "w")
l1 = l_x[0:]
for i in range(len(l_y)): l1.append(l_y[i])
f.write(str(l1))
f.close()
```
