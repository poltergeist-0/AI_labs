```python
import matplotlib.pyplot as plt
import numpy as np
import math

xx = np.linspace(-1,1,100)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)

plt.xlim(-1,1.5)
plt.ylim(-0.2,1.8)
plt.plot(xx,(abs((1-xx**2))**0.5),"black",label='y = |(1-x^2)^0.5|')
plt.legend(fontsize=11)
plt.show()

x = np.linspace(0,3,100)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)

plt.ylim(0,1)
plt.xlim(1,3)
plt.plot(x,((math.e)**(-x**2)),"black",label='y = exp(-x^2)')
plt.legend(fontsize=11)
plt.show()
```

```python
import scipy
from scipy import integrate
import math

def function_test(x):
    return (1 - x**2) ** 0.5
    
result_0 = round(integrate.quad(function_test,-1.0,1.0)[0], 5)
print("test result 1",result_0)
```

```python
def function_0(x):
    return math.exp(-x**2)
```

```python
import random
n = 25000 # -   

xt = np.linspace(-1,1,100)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)

l_x,l_y = [],[]
for i in range(n):
    xi,yi = random.uniform(-1.0,1.0),random.uniform(0.0,1.0)
    l_x.append(xi)
    l_y.append(yi)

points_below_x = []
points_below_y = []
points_above_x = []
points_above_y = []
n_below = 0

for i in range(n):
    if function_test(l_x[i]) > l_y[i]:
        points_below_x.append(l_x[i])
        points_below_y.append(l_y[i])
        n_below += 1
    else:
        points_above_x.append(l_x[i])
        points_above_y.append(l_y[i])

plt.plot(points_below_x,points_below_y,"og",markersize = 1)
plt.plot(points_above_x,points_above_y,"oy",markersize = 1)
plt.plot(xt,(abs((1-xt**2))**0.5),color="red",linewidth=1,label='y = |(1-x^2)^0.5|')

result_1 = (n_below/n) * 2
print("test result 2",result_1)

absolute_error = round(abs(result_0-result_1),5)
relative_error = round(abs(absolute_error/result_1),5)
print("absolute error",absolute_error)
print("relative error",relative_error)
#m1 = (absolute_error/(result_1))/2
#m2 = (relative_error/(result_1))/2
m1 = absolute_error/2
m2 = relative_error/2
#print(m1,m2)

plt.xlim(-1,1)
plt.ylim(0,1.5)
plt.legend(fontsize=11)
plt.show()
```

```python
X = np.linspace(1,3,100)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)

l_x,l_y = [],[]
maxf = function_0(1)
for i in range(n):
    xi,yi = random.uniform(1.0,3.0),random.uniform(0.0,maxf)
    l_x.append(xi)
    l_y.append(yi)

points_below_x = []
points_below_y = []
points_above_x = []
points_above_y = []
n_below = 0
for i in range(n):
    if function_0(l_x[i]) > l_y[i]:
        points_below_x.append(l_x[i])
        points_below_y.append(l_y[i])
        n_below += 1
    else:
        points_above_x.append(l_x[i])
        points_above_y.append(l_y[i])

result = round( n_below/n * maxf*(3-1) ,5)
print("result",result)
#print("absolute error estimate",round( result*m1 * maxf*2 ,5))
#print("relative error estimate",round( result*m2 * maxf*2 ,5))
print("absolute error estimate",round( m1 * maxf*2 ,5))
print("relative error estimate",round( m2 * maxf*2 ,5))

plt.plot(points_below_x,points_below_y,"og",markersize = 1)
plt.plot(points_above_x,points_above_y,"oy",markersize = 1)
plt.plot(X,((math.e)**(-X**2)),color="red",linewidth=1,label='y = exp(-x^2)')
plt.xlim(1,3)
plt.ylim(0,maxf)

plt.legend(fontsize=11)
plt.show()
```
